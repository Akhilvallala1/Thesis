````python
import re
import csv
from PyPDF2 import PdfReader

# Function to extract questions, options, and answers from the PDF
def extract_questions_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    full_text = ""

    # Combine text from all pages
    for page in reader.pages:
        full_text += page.extract_text()

    # Split text into questions and answers
    questions_pattern = r"\d+\)\s(.*?\?)\s(A\).+?D\).+?)(?=\n\d+\)|Answer Key)"
    matches = re.findall(questions_pattern, full_text, re.DOTALL)

    # Extract answers
    answer_pattern = r"\d+\)\s([A-D])"
    answers = re.findall(answer_pattern, full_text)

    questions_data = []

    for idx, (question, options) in enumerate(matches):
        questions_data.append({
            "Id": idx + 1,
            "Question": question.strip(),
            "Options": options.strip(),
            "Answer": answers[idx] if idx < len(answers) else ""
        })

    return questions_data

# Function to save extracted data into a CSV file
def save_to_csv(data, csv_path):
    with open(csv_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=["Id", "Question", "Options", "Answer"])
        writer.writeheader()
        writer.writerows(data)

# Path to the PDF file
pdf_path = "1dkin.pdf"

# Path to save the CSV file
csv_path = "output.csv"

# Extract questions and save them to CSV
questions_data = extract_questions_from_pdf(pdf_path)
save_to_csv(questions_data, csv_path)

print(f"Data has been successfully extracted and saved to {csv_path}.")
````

#Question pdf to json

````python
import openai
from langchain.document_loaders import PyPDFLoader
import csv
import json
import re

# Function to extract data from PDFs using OpenAI GPT
def extract_data_with_gpt(content, task):
    openai.api_key = "your_open_ai_key"  # Replace with your OpenAI API key

    messages = [
        {"role": "system", "content": "You are a helpful assistant that extracts structured data."},
        {"role": "user", "content": f"Extract the following information as structured JSON:\nTask: {task}\nContent: {content}"}
    ]

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=messages,
        max_tokens=1500,
        temperature=0
    )
    raw_response = response["choices"][0]["message"]["content"].strip()

    # Remove any Markdown formatting like ```json or ```
    cleaned_response = re.sub(r"```[a-zA-Z]*\n|```", "", raw_response)

    try:
        # Attempt to parse the cleaned JSON response
        return json.loads(cleaned_response)
    except json.JSONDecodeError:
        print("Error: Unable to parse GPT response as JSON.\nResponse:", cleaned_response)
        return []

# Function to parse questions PDF
def parse_questions_pdf(file_path):
    loader = PyPDFLoader(file_path)
    pages = loader.load()
    questions = []

    for page_number, page in enumerate(pages, start=1):
        print(f"Processing page {page_number}...")
        content = page.page_content.strip()
        print(f"Page content:\n{content[:500]}")  # Print the first 500 characters of the page

        structured_data = extract_data_with_gpt(
            content,
            "Extract a list of questions with the following JSON format: [{'Id': <id>, 'Question': <question>, 'Options': <options>}]"
        )

        if structured_data:
            print(f"Extracted data from page {page_number}: {structured_data}")
            questions.extend(structured_data)
        else:
            print(f"No structured data extracted from page {page_number}. Raw content:\n{content}")

    return questions




# Main execution
questions_pdf_path = "Questions.pdf"  # Path to your questions PDF


# Process PDFs
questions = parse_questions_pdf(questions_pdf_path)
````

#Answer pdf to dataframe

````python
import openai
from langchain.document_loaders import PyPDFLoader
import pandas as pd
import re
import json

# Function to extract answers from PDFs using LangChain and OpenAI
def parse_answers_pdf_with_langchain(pdf_path):
    loader = PyPDFLoader(pdf_path)
    pages = loader.load()

    answers = []

    for page_number, page in enumerate(pages, start=1):
        print(f"Processing page {page_number}...")
        content = page.page_content.strip()

        # Use OpenAI to parse answers
        openai.api_key = "your_open_api_key"  # Replace with your OpenAI API key
        messages = [
            {"role": "system", "content": "You are a helpful assistant that extracts structured data."},
            {"role": "user", "content": f"Extract the answers in the following JSON format: [{{\"Id\": <id>, \"Answer\": <answer>}}]. The IDs are numeric and answers are single characters (A, B, C, D).\nContent:\n{content}"}
        ]

        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=messages,
            max_tokens=1500,
            temperature=0
        )

        raw_response = response["choices"][0]["message"]["content"].strip()

        # Remove any Markdown formatting like ```json
        cleaned_response = re.sub(r"```[a-zA-Z]*\n|```", "", raw_response)

        try:
            # Parse the JSON response safely
            parsed_data = json.loads(cleaned_response)
            answers.extend(parsed_data)
        except json.JSONDecodeError as e:
            print(f"Error parsing OpenAI response as JSON: {e}\nResponse: {cleaned_response}")
        except Exception as e:
            print(f"Unexpected error: {e}\nResponse: {cleaned_response}")

    # Convert answers to DataFrame
    answers_df = pd.DataFrame(answers)
    return answers_df

# Path to the answers PDF
answers_pdf_path = "Answers.pdf"

# Parse the PDF to extract answers
answers_df = parse_answers_pdf_with_langchain(answers_pdf_path)

# Save the extracted data to a CSV
output_csv_path = "answers_dataset_langchain.csv"
answers_df.to_csv(output_csv_path, index=False)

print(f"Answers dataset saved to {output_csv_path}")

# Display the DataFrame
answers_df.head()
````

#Data Cleaning

````python
import ast

# Function to convert the options from list format to dictionary format
def convert_options(option_list):
    if isinstance(option_list, str):  # Ensure the column has stringified lists
        try:
            option_list = ast.literal_eval(option_list)  # Convert to list
            return {item.split(') ')[0]: item.split(') ')[1] for item in option_list}
        except (ValueError, SyntaxError):
            return option_list  # Return original value if conversion fails
    return option_list

# Apply the conversion to the 'Options' column
merged_df['Options'] = merged_df['Options'].apply(convert_options)
````

````python
# Function to map list to dictionary with A, B, C, D keys
def add_abcd(option_list):
    if isinstance(option_list, list):  # Check if it's a list
        keys = ['A', 'B', 'C', 'D']  # Define keys
        return {keys[i]: option_list[i] for i in range(len(option_list))}
    return option_list  # Return original value if not a list

# Apply the conversion to the 'Options' column
merged_df['Options'] = merged_df['Options'].apply(add_abcd)
````
